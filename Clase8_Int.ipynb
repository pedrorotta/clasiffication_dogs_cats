## ¿Cómo cambiar el tamaño de una imágen?

Esto es importante, porque las imágenes no siempre van a estar en un tamaño igual, a veces tendremos imágenes de diferentes tamaños y tenemos que convertirlas todas al mismo tamaño

```python
# elegimos un tamaño y lo guardamos en una tupla

tamaño = (300,300)
gato3.thumbnail(tamaño,)
gato3.save("gato2.png")
```

```python
#importamos gato (la imagen original)
```

```python
Image.open("/content/gato.png")
```

```python
Image.open("/content/gato2.png")
```
